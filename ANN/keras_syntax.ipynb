```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/fake_reg.csv') # czytanie pliku csv
```

```python
df.head() # matrix pliku csv
```

```python
sns.pairplot(df) # plot par
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values # .values od tablicy z pandas
```

```python
y = df['price'].values # .values od tablicy z pandas
```

```python
# Split arrays or matrices into random train and test subsets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) 
```

```python
X_train
```

```python
X_train.shape # pandas_matrix.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() # scalowanie w do jakiegos given range 0-1 
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train.min()
```

```python
X_train.max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# model = Sequential([Dense(4, activation='relu'), Dense(2, activation = 'relu'), Dense(1)])
```

```python
# model sekwencyjny 4 warstwowy z  funkcja aktywacyjna relu
model = Sequential()
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot() # wykres strat
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions
```

```python
test_predictions = pd.Series(test_predictions.reshape(330,))
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis=1)
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
```

```python
sns.scatterplot(x='Test True Y',y='Model Predictions',data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'], pred_df['Model Predictions'])
```

```python
# df.describe()
```

```python
mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions'])
```
